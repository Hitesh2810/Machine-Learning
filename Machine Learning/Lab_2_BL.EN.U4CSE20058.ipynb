```python
#A1
import nltk
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
Sentence=pd.read_csv("F:\\Amrita\\5th Semester\\Machine Learning\\Lab\\Lab-2\\Lab_2.csv")

print(Sentence)
```

```python
#A1
Word_tokenise=[]
for index,row in Sentence.iterrows():
  Word_tokenise.append(word_tokenize(row['Report'].lower()))
print(Word_tokenise)
  
```

```python
#A2
Master_list=[]
for i in Word_tokenise:
    for j in i:
        if j not in Master_list:
            Master_list.append(j)
print("The Distinct tokens : \n",Master_list)
```

```python
#A3
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
#A4
Bag_of_words=[]
for i in Word_tokenise:
    for j in i:
        if j not in Bag_of_words: 
            if j not in stopwords.words('english'):
                Bag_of_words.append(j)
print("The Bag of words: \n",Bag_of_words)
```

```python
#A5 Creating Vector-1
Vector_1 = []
for i in Word_tokenise:
  temp = []
  for m in Bag_of_words:
    if(m in i):
      temp.append(1)
    else:
      temp.append(0)
  Vector_1.append(temp)
for i in Vector_1:
  print (i)

#A5 Creating Vector-2
Vector_2 = []
for j in Word_tokenise:
  temp = []
  for n in Bag_of_words:
    temp.append(j.count(n))
  Vector_2.append(temp)
for j in Vector_2:
  print(j)
 
```

```python
#A6
def Jacc_Coeff(a, b):
  f11 = 0
  f10 = 0
  f01 = 0
  for i in range(len(Bag_of_words)):
    if (a[i] == 1 and b[i] == 1):
         f11+=1
    elif (a[i] == 0 and b[i] == 1):
        f01 += 1
    elif (a[i] == 1 and b[i] == 0):
        f10 += 1
  return (f11/(f11+f01+f10))

Jaccard_matrix = []
for i in range(len(Vector_1)):
  temp = []
  for j in range(len(Vector_1)):
    print("The similarity between {} and {} documents is".format(i,j),Jacc_Coeff(Vector_1[i],Vector_1[j]))
    temp.append(Jacc_Coeff(Vector_1[i],Vector_1[j]))
  Jaccard_matrix.append(temp)

print('\nThe plagarism percent of documents 32 and 14 using jaccard coefficient is: ')
print((Jaccard_matrix[32][14])*100,'%')
```

```python
#A7
def Cos_sim(a, b):
  return (np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b)))

Cosine_matrix = []
for i in range(len(Vector_2)):
  temp = []
  for j in range(len(Vector_2)):
    print("The similarity between {} and {} documents is".format(i,j),Cos_sim(Vector_2[i],Vector_2[j]))
    temp.append(Cos_sim(Vector_2[i],Vector_2[j]))
  Cosine_matrix.append(temp)
print('\nThe plagarism percent of documents 32 and 14 using cosine similarity is: ')
print((Cosine_matrix[32][14])*100,'%')
```
